```python
import pandas as pd

# Data
data = {
    'Student': ['Grad_1', 'Grad_2', 'Grad_3', 'Grad_4', 'Grad_5', 'Grad_6', 'Grad_7', 'Grad_8', 'Grad_9', 'Grad_10', 'Grad_11', 'Grad_12', 'Grad_13', 'Grad_14', 'Grad_15'],
    'Salary': [1550, 2500, 2645, 7000, 15000, 3700, 2530, 2050, 2200, 2050, 2400, 2350, 2700, 4500, 3300],
    'Final_grade': [6.5, 7.0, 8.5, 7.0, 7.0, 6.0, 8.0, 7.0, 6.0, 7.0, 8.0, 7.5, 6.0, 7.0, 7.5]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate statistics
mean = df['Salary'].mean()
median = df['Salary'].median()
std_dev = df['Salary'].std(ddof=1)  # ddof=1 for sample standard deviation

print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Sample Standard Deviation: {std_dev}")
```

```python
import matplotlib.pyplot as plt
# Calculate correlation
correlation = df['Salary'].corr(df['Final_grade'])

print(f"Correlation: {correlation}")
```

```python
# Create scatterplot
plt.scatter(df['Final_grade'], df['Salary'])
plt.title('Salary vs Final Grade')
plt.xlabel('Final Grade')
plt.ylabel('Salary')
plt.show()
```

```python
import seaborn as sns

# Create a scatterplot with regression line
sns.regplot(x='Final_grade', y='Salary', data=df)
plt.title('Correlation between Salary and Final Grade')
plt.xlabel('Final Grade')
plt.ylabel('Salary')
plt.show()
```
